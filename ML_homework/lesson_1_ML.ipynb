Задание:
Предсказать параметр MEDV из датасета housing.csv:
1) модель линейной регрессии

2) разбить train/test на 0.8/0.2

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

RANDOM_SEED = 1
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
#делим выборку
X_train, X_test, y_train, y_test = train_test_split(df.drop('MEDV', axis = 1), 
                                                    df['MEDV'], 
                                                    test_size = 0.2,
                                                   random_state = RANDOM_SEED)
```

```python
#обучаем модель
model = LinearRegression()

model.fit(X_train,y_train)

pred = model.predict(X_test)
```

```python
#добавим столбец
pred = model.predict(df.drop('MEDV', axis = 1))
df['predict'] = pred
```

```python
#построим график для визуального сравнения
df[['MEDV','predict']].plot(figsize=(30, 10))
```

Модель визуально повторяет контуры фактических данных.
